```python
from __future__ import annotations
import types
import copy

class GenericTest[T: NamedTuple]:
    TYPE_VAR: type[T]

    def __new__(cls: type[GenericTest[T]]) -> GenericTest[T]:
        params = dict(copy.deepcopy(inspect.signature(cls.f)).parameters)
        for field in cls.TYPE_VAR._fields:
            params[field] = Parameter(
                name=field,
                kind=Parameter.POSITIONAL_OR_KEYWORD,
                annotation=cls.TYPE_VAR.__annotations__[field],
            )

        print(params)
        del params["args"]

        signature._parameters = types.MappingProxyType(params)
        cls.f.__signature__ = signature
        return cls()
    
    def __init__(self) -> None:
        pass

    def f(self, *args: Any) -> T:
        print(args)
        return self.TYPE_VAR._make(args)
```
