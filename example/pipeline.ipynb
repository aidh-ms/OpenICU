```python
from pathlib import Path

import polars as pl

from open_icu.steps.concept.config.concept import ConceptConfig

c = ConceptConfig.load(Path("/workspaces/config/concept/medication/antibiotics.yml"))
pattern = c.mappings[0].regex
print(pattern)

lf = pl.scan_parquet(project.datasets["extraction"].metadata_path / "codes.parquet")
lf.filter(pl.col("code").str.contains(pattern)).collect()["code"].str.to_lowercase().to_list()
```
